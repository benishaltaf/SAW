```python

import random
import matplotlib.pyplot as plt

moves = [
    (1, 0),   # right
    (-1, 0),  # left
    (0, 1),   # up
    (0, -1)   # down
]

# Starting point
x, y = 0, 0

visited = set()
visited.add((x, y))

path_x = [x]
path_y = [y]

steps = 200

for _ in range(steps):

    # Find allowed moves
    possible_moves = []

    for dx, dy in moves:
        nx, ny = x + dx, y + dy

        if (nx, ny) not in visited:
            possible_moves.append((nx, ny))

    if not possible_moves:
        print("Walker trapped!")
        break

    x, y = random.choice(possible_moves)

    # Mark visited
    visited.add((x, y))

    # Save path
    path_x.append(x)
    path_y.append(y)

# Plot
plt.figure(figsize=(6,6))
plt.plot(path_x, path_y, marker='o')
plt.title("Self-Avoiding Random Walk")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.axis('equal')

plt.show()
```
